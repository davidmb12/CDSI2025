```python
import librosa
from IPython.display import Audio
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf

audio, sr = librosa.load("DanielCaesarSong.mp3")
mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
```

```python
# Compute the spectrogram
spectrogram = librosa.amplitude_to_db(librosa.stft(audio), ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(spectrogram, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```

```python
# Mel Spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr)
librosa.display.specshow(librosa.power_to_db(mel_spectrogram, ref=np.max), sr=sr, x_axis='time', y_axis='mel')
plt.colorbar()
plt.title("Mel Spectrogram")
plt.show()
```

```python
# Tempo 
tempo, beats = librosa.beat.beat_track(y=audio, sr=sr)
print(f"Tempo: {tempo} BPM")
```

```python
zero_crossing_rate = librosa.feature.zero_crossing_rate(audio)
print(zero_crossing_rate)
```

```python
mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)

# Plot the MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.show()
```

```python
## -------- PyTorch
```

```python
import torch
import torchaudio
import matplotlib.pyplot as plt

# Load an audio file
waveform, sample_rate = torchaudio.load("DanielCaesarSong.mp3")

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(waveform.t().numpy())
plt.title(f"Waveform (Sample Rate: {sample_rate})")
plt.show()
```

```python
resampler = torchaudio.transforms.Resample(orig_freq=44100, new_freq=16000)

resampled_waveform = resampler(waveform)

print("Original Shape:", waveform.shape)
print("Resampled Shape:", resampled_waveform.shape)
```

```python
transform = torchaudio.transforms.MelSpectrogram(sample_rate=sample_rate, n_mels=64)
mel_spectrogram = transform(waveform)

# Mel spectrogram
plt.figure(figsize=(10, 4))
plt.imshow(mel_spectrogram.log2()[0].numpy(), aspect="auto", origin="lower")
plt.title("Mel Spectrogram")
plt.colorbar(label="Log Power")
plt.show()
```

```python
mfcc_transform = torchaudio.transforms.MFCC(sample_rate=sample_rate, n_mfcc=13)
mfccs = mfcc_transform(waveform)

print("MFCCs Shape:", mfccs.shape)
```

```python
## --------- PY Audio
```

```python
import pyaudio
import wave

# Parameters
chunk = 1024  # Buffer size
format = pyaudio.paInt16  # Audio format (16-bit PCM)
channels = 1  # Number of channels (1 for mono, 2 for stereo)
rate = 44100  # Sampling rate (Hz)
record_seconds = 5  # Duration of recording
output_filename = "output.wav"

# Initialize PyAudio
p = pyaudio.PyAudio()

# Open a stream for recording
stream = p.open(format=format, channels=channels,
                rate=rate, input=True,
                frames_per_buffer=chunk)

print("Recording...")

frames = []

# Capture audio data
for _ in range(0, int(rate / chunk * record_seconds)):
    data = stream.read(chunk)
    frames.append(data)

print("Recording finished.")

# Stop and close the stream
stream.stop_stream()
stream.close()
p.terminate()

# Save the audio to a WAV file
with wave.open(output_filename, 'wb') as wf:
    wf.setnchannels(channels)
    wf.setsampwidth(p.get_sample_size(format))
    wf.setframerate(rate)
    wf.writeframes(b''.join(frames))

print(f"Audio saved to {output_filename}")
```

```python
## -------- SciPy
```

```python
from scipy.io import wavfile
import numpy as np
from scipy.signal import butter, lfilter
import matplotlib.pyplot as plt
from scipy.fft import fft
# Read a WAV file
sample_rate, data = wavfile.read("Daniel-Caesar-Best-Part-A-COLORS-SHOW.wav")
print(f"Sample Rate: {sample_rate} Hz")
print(f"Data Shape: {data.shape}")
```

```python
# Compute FFT
N = len(data)
frequencies = np.fft.fftfreq(N, d=1/sample_rate)
fft_values = fft(data)

# Plot the magnitude spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:N // 2], np.abs(fft_values)[:N // 2])
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```

```python

# Design a low-pass filter
def low_pass_filter(data, cutoff_freq, sample_rate):
    nyquist = 0.5 * sample_rate
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(5, normal_cutoff, btype='low', analog=False)  # 5th order filter
    return lfilter(b, a, data)

# Apply the filter
cutoff_frequency = 5000  # 5 kHz
filtered_data = low_pass_filter(data, cutoff_frequency, sample_rate)

# Save the filtered data
wavfile.write("low_pass_output.wav", sample_rate, filtered_data.astype(np.int16))
```

```python
from scipy.signal import convolve

# Simple reverb effect
impulse_response = np.zeros(2000)
impulse_response[0] = 1
impulse_response[500] = 0.5
impulse_response[1000] = 0.3

# Convolve the signal with the impulse response
reverb_data = convolve(data, impulse_response, mode='full')

# Save the audio with reverb
wavfile.write("reverb_output.wav", sample_rate, reverb_data.astype(np.int16))
```
