```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, exposure
import PIL.Image as Image
from scipy.signal import convolve2d

from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
```

```python
def initialize_means(img,clusters):
    points = img.reshape((-1,img.shape[2]))
    m,n = points.shape

    means = np.zeros((clusters,n))
    for i in range(clusters):
        rand_indices = np.random.choice(m,size = 10,replace = False)
        means[i] = np.mean(points[rand_indices],axis = 0)
    return points,means

def distance(x1,y1,x2,y2):
    dist = np.square(x1-x2) + np.square(y1-y2)
    dist = np.sqrt(dist)
    return dist

def k_means(points,means,clusters):
    iterations = 10
    m,n = points.shape

    index = np.zeros(m)
    while iterations > 0:
        for j in range(m):
            min_dist = float('inf')
            temp = None
            for k in range(clusters):
                x1,y1 = points[j,0],points[j,1]
                x2,y2= means[k,0],means[k,1]
                if distance(x1,y1,x2,y2) <= min_dist:
                    min_dist = distance(x1,y1,x2,y2)
                    temp = k
                    index[j] = k
        for k in range(clusters):
            cluster_points = points[index==k]
            if len(cluster_points)>0:
                means[k] = np.mean(cluster_points,axis = 0)
        iterations -=1
    return means,index

def compress_image(means,index,img):
    centroid = np.array(means)
    recovered = centroid[index.astype(int),:]

    recovered = recovered.reshape(img.shape)
    plt.imshow(recovered)
    plt.show()
    cv2.imwrite('compressed_'+str(clusters) +'_colors.png',recovered)
    
```

```python
def show_image(title, img):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def load_image(path):
    return cv2.imread(path, cv2.IMREAD_COLOR)
```

<h2>Compressing Images</h2>

```python
img_01 = cv2.imread(image_path_01)  # Replace with the actual path to your image
# Convert the image from BGR to RGB
img_01 = cv2.cvtColor(img_01, cv2.COLOR_BGR2RGB)
     
# Scaling the image so that the values are in the range of 0 to 1
img_01 = img_01 / 255.0
clusters = 16
clusters = int(input("Enter number of colors in the compressed image"))

points,means= initialize_means(img_01,clusters)
means,index = k_means(points,means,clusters)
compress_image(means,index,img_01)
```

<h2>Correccion Gamma
</h2>

```python
def adjust_gamma(image,gamma=1.0):
    invGamma = 1.0/gamma
    table = np.array([((i/255.0)**invGamma)*255 for i in range(256)]).astype("uint8")
    return cv2.LUT(image,table)

image = cv2.imread(image_path_02)
gamma_corrected = adjust_gamma(image,gamma = 2.0)
cv2.imshow("Original",image)
cv2.imshow("Gamma Corregido",gamma_corrected)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
while cap.isOpened():
    ret,frame = cap.read()
    if not ret:
        break
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    keypoints,descriptors = orb.detectAndCompute(gray,None)
    frame_keypoints = cv2.drawKeypoints(frame,keypoints,None,color = (0,255,0))
    
    cv2.imshow("ORB Features",frame_keypoints)
    if cv2.waitKey(1)& 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model

# Load the pre-trained VGG16 model and choose a layer for feature extraction
base_model = VGG16(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc1').output)

# Open the video file
cap = cv2.VideoCapture("./Images/VideoLala.mp4")

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Resize frame to 224x224 (VGG16 input size) and convert BGR (OpenCV) to RGB
    frame_resized = cv2.resize(frame, (224, 224))
    frame_rgb = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2RGB)
    
    # Prepare the frame for prediction
    x = image.img_to_array(frame_rgb)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    # Extract features
    features = model.predict(x)
    print(f"Frame {frame_count} feature vector shape: {features.shape}")
    
    frame_count += 1
    # For demonstration, process only the first few frames
    if frame_count >= 10:
        break

cap.release()
```
