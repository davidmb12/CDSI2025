<h2>Importacion de librerias necesarias</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

from audioLoading import loadAudios
from scipy.io import wavfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
from convertFiles import convertFiles
from pathlib import Path
from scipy.io import wavfile
```

Preparamos un diccionario que relaciona los nombres en los archivos de audio con un indice unico que funcionara como un label

```python
#
audioToIdx ={
    'AdriaM':0,
    'AleM':1,
    'BetoM':2,
    'Cielo':3,
    'Daniel':4,
    'Erik':5,
    'Irma':6,
    'LuisG':7,
    'Maria':8,
    'Mariana':9,
    'Mario':10,
    'Marlon':11,
    'MauM':12,
    'Sergio':13,
    'Vanessa':14,
    'David':15,
    'Nula':16,
}
```

Cargar los audios

```python
audios = loadAudios(f"./AUDIO/All-Audios/*")
audios_features = []
```

```python
print(audios)
```

Extraccion de caracteristicas

```python
audios_features = []
for audio_signal,filename in audios:
    features = [
        librosa.feature.spectral_contrast(y=audio_signal[0])[0,0],
        librosa.feature.chroma_cens(y=audio_signal[0])[0,0],
        librosa.feature.mfcc(y=audio_signal[0])[0,0],
    ]
    label = audioToIdx[(str.split(filename,'-')[0])]
    audios_features.append([label]+features)
```

```python
print(audios_features)
```

Normalizar datos

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(audios_features)
print(features_scaled)
```

Crear el DataFrame

```python
df = pd.DataFrame(features_scaled,columns=["Label","Spectral Contrast","Chroma","MFCC"])
df.to_csv("audio_features.csv",index=False)
```

```python
print(df)
```

Dividir los datos en conjuntos de entrenamiento y prueba

```python
x = df[["Spectral Contrast","Chroma","MFCC"]]
y = df["Label"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

Crear y entrenar el modelo (Decision Tree)

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

Evaluar el modelo (Decision Tree)

```python
accuracy = clf.score(x_test,y_test)
print(accuracy)
```

Crear y entrenar modelo (KNeighborsClassifier)

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
accuracy_knn = knn.score(x_test,y_test)
print(accuracy_knn)
```
